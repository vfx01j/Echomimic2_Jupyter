```python
from google.colab import files
import os

# Define the path to your output folder
output_folder = "/content/echomimic_v2/output/"

# Find the most recently generated mp4 file in that folder
mp4_files = [f for f in os.listdir(output_folder) if f.endswith('.mp4')]

if mp4_files:
    # Sort files to get the newest one
    mp4_files.sort(key=lambda x: os.path.getmtime(os.path.join(output_folder, x)))
    latest_video = mp4_files[-1]
    full_path = os.path.join(output_folder, latest_video)

    print(f"Downloading: {latest_video} to your Mac...")
    files.download(full_path)
else:
    print("No generated MP4 videos found in the output folder yet.")
```
